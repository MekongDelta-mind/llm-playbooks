```python
import pandas as pd
import os
import kagglehub
from sklearn.metrics import precision_score
from sentence_transformers import SentenceTransformer
from sentence_transformers import SentenceTransformerTrainer, losses
```

```python
path = kagglehub.dataset_download("rmisra/news-category-dataset")

json_file = os.path.join(path, 'News_Category_Dataset_v3.json')  # Adjust filename if needed
df = pd.read_json(json_file, lines=True)

df.head()
```

```python
# load model and create embedding for headline
model = SentenceTransformer("nomic-ai/nomic-embed-text-v1.5", trust_remote_code=True)
headline = df.loc[0, 'headline']
headline_embedding = model.encode(headline)

# create other embeddings
embeddings = model.encode(df.loc[1:100,'headline'].tolist())
```

```python
results = []
for truncation_size in [768, 512, 256, 128]:
  similarities = model.similarity(headline_embedding[:truncation_size], embeddings[:,:truncation_size])
  retrieved_scores = similarities.sort(descending=True)[0][:25].tolist()
  top_25_indices = similarities.argsort(descending=True)[0][:25].tolist()
  top_25_sentences = [df.at[idx, 'headline'] for idx in top_25_indices]
  top_25_df = pd.DataFrame(top_25_sentences, columns=["Top 25 Headlines"])
  top_25_df['truncation_size'] = truncation_size
  results.append(top_25_df)

top_25_df = pd.concat(results)

top_25_df
```

```python
# Notes to reader: filter top_25_df for each truncation size, and manually fill in labels as per your judgement of relevance
print(top_25_df.loc[top_25_df['truncation_size']==768, 'Top 25 Headlines'])
relevance_labels = [1, 0, 1, 1, ...]  # Replace with actual relevance values

# Create DataFrame to visualize results
results_df = pd.DataFrame({
    'Sentence': top_25_sentences,
    'Cosine_Score': retrieved_scores,
    'Relevance': relevance_labels
})

print(results_df)

precision_at_25 = precision_score(relevance_labels, [1]*25)
print(f"Precision@25: {precision_at_25}")


```

```python
from datasets import Dataset

mat_model = SentenceTransformer("nomic-ai/nomic-embed-text-v1.5", trust_remote_code=True)
loss = losses.MultipleNegativesRankingLoss(mat_model)
loss = losses.Matryoshka2dLoss(mat_model, loss, [768, 512, 256, 128, 64])

dataset = Dataset.from_pandas(df[['headline']].iloc[:1000])

trainer = SentenceTransformerTrainer(
    model=mat_model,
    train_dataset=dataset,
    loss=loss,
)
trainer.train()
```

```python
from qdrant_client.models import VectorParams, Distance

collection_name = "headline_embeddings_768"

if not qdrant_client.collection_exists(collection_name):
    qdrant_client.create_collection(
        collection_name=collection_name,
        vectors_config=VectorParams(size=768, distance=Distance.COSINE),
    )
```

```python
points = [{"id": id, "vector": embedding} for id, embedding in enumerate(embeddings)]

# Upsert points into Qdrant
qdrant_client.upsert(collection_name=collection_name, points=points)
```

```python
search_results = qdrant_client.search(
    collection_name=collection_name,
    query_vector=headline_embedding,
    limit=5  # Adjust this number based on how many similar embeddings you want to retrieve
)

for result in search_results:
    print(f"ID: {result.id}")
    print(f"Sentence: {df.loc[result.id,'headline']}")
    print(f"Payload: {result.payload}")
    print(f"Score: {result.score}")
    print("-" * 20)
```
